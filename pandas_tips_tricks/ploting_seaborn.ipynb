# Box Plot

```python
df['class'].unique()
```

```python
# sns set style or theme
# style of plot
sns.set_theme(style="darkgrid",font_scale=1.5,font='Times New Roman')
# plot size
plt.figure(figsize=(12, 8))
# change the scatter point colors
palette={'First': 'blue', 'Second': 'green','Third':"pink"}
sns.boxplot(x="sex", y="age", data=df, hue='class', palette=palette)

# labels
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Box plot of age vs fare")
# change legend position and title
plt.legend(loc='upper right', title='Class Name')
```

```python
# sns set style or theme
# style of plot
sns.set_theme(style="whitegrid",font_scale=1.5,font='Times New Roman')
# plot size
plt.figure(figsize=(12, 8))
# change the scatter point colors
palette={'First': 'blue', 'Second': 'green','Third':"pink"}
sns.boxenplot(x="sex", y="age", data=df, hue='class', palette=palette)

# labels
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Box plot of Age,based on Gender and Class Name of Passengers")
# change legend position and title
plt.legend(loc='upper right', title='Class Name')
# change the x tick labels
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
# rotate x ticks labels by 45 degree
#plt.xticks(rotation=45)
```

```python
# make scatter plot with plotly
import plotly.express as px
fig=px.scatter(df,x='age',y='fare',color='sex')
fig.show()
```

```python
# bar plot using seaborn
# sns set style or theme
# style of plot
sns.set_theme(style="whitegrid",font_scale=1.5,font='Times New Roman')
# plot size
plt.figure(figsize=(12, 8))
# change the scatter point colors
palette={'First': 'blue', 'Second': 'green','Third':"pink"}
sns.barplot(x="sex", y="age", data=df, hue='class', palette=palette,capsize=0.5)

# labels
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Box plot of Age,based on Gender and Class Name of Passengers")
# change legend position and title
plt.legend(loc='upper right', title='Class Name')
# change the x tick labels
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
```

```python
# sns set style or theme
# style of plot
sns.set_theme(style="whitegrid",font_scale=1.5,font='Times New Roman')
# plot size
plt.figure(figsize=(12, 8))
# change the scatter point colors
palette={'First': 'blue', 'Second': 'green','Third':"pink"}
sns.violinplot(x="sex", y="age", data=df, hue='class', palette=palette)

# labels
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Box plot of Age,based on Gender and Class Name of Passengers")
# change legend position and title
plt.legend(loc='upper right', title='Class Name')
# change the x tick labels
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
# rotate x ticks labels by 45 degree
```
